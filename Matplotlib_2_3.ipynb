```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
fig, axes = plt.subplots(1,3)
# for i in axes:
#     i.plot(x,y)
axes[0].plot(x, y)
axes[0].set_title('first plot')
axes[1].plot(y, 2*x)
axes[1].set_title('second plot')
plt.tight_layout()
```

```python
axes
```

```python
fig = plt.figure(figsize = (8, 2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=300)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, 'r-',label='X', color='purple')
ax.plot(y, x, 'g-',label='Y', color ='#FF8C00', linestyle='--', marker='o', markersize=10, markerfacecolor='pink', markeredgewidth=3)
ax.plot(x, 2*x**2 ,label='2X squared', color='black', linestyle='steps')
ax.plot(x, x**3,label='X cubed', color='green', linewidth=5 , alpha = 0.5, linestyle='-.')

ax.set_title('Title')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend(loc = 0)
```
